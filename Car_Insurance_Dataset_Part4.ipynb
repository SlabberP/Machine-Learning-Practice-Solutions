##**Preprocessing**

### **Create a Copy of the Dataset**

```python
# Create a copy of the dataset for Modeling
ml_df = df.copy()
```

### **Check Class Balance**

```python
# Check how many samples of each class are present
ml_df['OUTCOME'].value_counts(normalize=False)
```

### **Split the Data**

```python
## Define features (X) and target (y)
target = 'OUTCOME'
X = ml_df.drop(columns = [target])
y = ml_df[target]
```

```python
ml_df.head()
```

**Please note that the split is different for classification. Check stratify=y when the model is not well balanced**

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
```

#### **Numeric Transformation**

```python
# We will use a mean imputer to impute missing values for the numeric columns
# Defining lists of types of features
num_cols = X_train.select_dtypes("number").columns
num_cols

# instantiate preprocessors
impute_mean = SimpleImputer(strategy='mean')
scaler = StandardScaler()

# Pipeline
num_pipe = make_pipeline(impute_mean, scaler)
num_pipe

# Tuple
numeric_tuple = (num_pipe, num_cols)
numeric_tuple

# Fit the pipeline on the numeric training data
num_pipe.fit(X_train[num_cols])
```

#### **Ordinal Transformation**

```python
# Ordered Category Lists

# define a list of columns to encode as ordinal
ordinal_cols = ['EDUCATION', 'INCOME', 'VEHICLE_YEAR']
ordinal_cols

# Specifying the order of categories in quality/condition columns
# impute_na_ord = SimpleImputer(strategy='constant', fill_value='NA')
impute_na_ord = SimpleImputer(strategy='most_frequent')

# Specifying order of categories for our  Ordinal Qual/Cond Columms
education_cond_order = ['high school','university']
income_cond_order = ['poverty', 'working class', 'middle class ', 'upper class']
vehicle_year_cond_order = ['before 2015', 'after 2015']

ord = OrdinalEncoder(categories=[education_cond_order, income_cond_order, vehicle_year_cond_order],handle_unknown='use_encoded_value',unknown_value = -1)

# Making a final scaler to scale category #'s
scaler_ord = StandardScaler()

# Pipeline
ord_pipeline = make_pipeline(impute_na_ord, ord, scaler_ord)

# Tuple
ordinal_tuple = (ord_pipeline, ordinal_cols)
ordinal_tuple

# Fit the pipeline on the training data
ord_pipeline.fit(X_train[ordinal_cols])
```

###**Nominal transformation**

```python
# Defining list of nominal features
ohe_cols = X_train.select_dtypes('object').drop(columns=ordinal_cols).columns
ohe_cols

# Instantiate OneHotEncoder
ohe_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Instantiate the pipeline
ohe_pipe = make_pipeline(ohe_encoder)
ohe_pipe

ohe_tuple = (ohe_pipe, ohe_cols)
ohe_tuple

# Fit the pipeline on the training data
ohe_pipe.fit(X_train[ohe_cols])
```

###**Instantiate Preprocessor**

```python
# Instantiate the make column transformer
preprocessor = make_column_transformer(numeric_tuple,
                                           ohe_tuple,
                                           ordinal_tuple,
                                       remainder='drop')
```

##**Models**


### **KNN Model**

```python
# Create an instance of the model
knn = KNeighborsClassifier()
# Create a model pipeline
knn_pipe = make_pipeline(preprocessor, knn)
## Fit the model
knn_pipe.fit(X_train, y_train)
```

### **KNN Model Evaluation**

```python
## Save the model's predictions for the train dataset
knn_train_preds = knn_pipe.predict(X_train)
## Save the model's predictions for the test dataset
knn_test_preds = knn_pipe.predict(X_test)
```

```python
## print the classification reports
labels = ['0','1']
print('Train Report')
print(classification_report(y_train, knn_train_preds, target_names=labels))

print('Test Report')
print(classification_report(y_test, knn_test_preds, target_names=labels))
```

```python
## create confusion matrices
ConfusionMatrixDisplay.from_predictions(y_train, knn_train_preds,
               normalize='true', display_labels=labels, cmap='Greens')
plt.suptitle('Train');



ConfusionMatrixDisplay.from_predictions(y_test, knn_test_preds,
               normalize='true', display_labels=labels, cmap='Greens')
plt.suptitle('Test');
```

The confusion Matrix shows the Type1 errors are lower during the training phase then the test phase.

The Type2 errors are higher during the training phase then the test phase.

Model did not perform great.

###**Tune the Hyperparameters**

```python
# Display the hyperparameters.
knn_pipe.get_params()
```

```python
# Create a dictionary of parameters

params = {'kneighborsclassifier__n_neighbors': range(1,15),
          'kneighborsclassifier__weights': ['distance','uniform']}
```

###**PCA**

```python
# Create a copy of the dataset for Modeling
ml_df = df.copy()

# Calculate the mean of the CREDIT_SCORE column
mean_value = ml_df['CREDIT_SCORE'].mean()
mean_value

# Replace null values with the mean value
ml_df['CREDIT_SCORE'].fillna(mean_value, inplace=True)

# Calculate the mean of the ANNUAL_MILEAGE column
mean_value2 = ml_df['ANNUAL_MILEAGE'].mean()
mean_value2

# Replace null values with the mean value
ml_df['ANNUAL_MILEAGE'].fillna(mean_value, inplace=True)

## Define features (X) and target (y)
target = 'OUTCOME'

df_X = ml_df.drop(columns = [target])
y = ml_df[target]

#Select numerical columns
X = df_X.select_dtypes(include=['number'])
X.head()

# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Perform PCA
n_components = 5  # Number of principal components to keep
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(scaled_data)

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)

# Create a transformer pipeline
transformer = make_pipeline(StandardScaler(), PCA(n_components=7))
```
